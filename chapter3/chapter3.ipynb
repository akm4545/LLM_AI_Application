# **카테고리를 문자로 표기한 label_str 컬럼 추가**

```python
klue_tc_train.features['label']
# 항목별 이름 확인
# ClassLabel(names=['IT과학', '경제', '사회', '생활문화', '세계', '스포츠', '정치'], id=None)

klue_tc_train.features['label'].int2str(1)
# 숫자형 아이디를 라벨 이름으로 변환
# '경제'

# 라벨만 추출
klue_tc_label = klue_tc_train.features['label']

def make_str_label(batch):
  # batch['label_str'] -> label_str 컬럼 추가
  # klue_tc_label.int2str(batch['label']) -> 각 데이터 별 label을 int2str을 통해 텍스트로 변환
  # features['lable'](라벨 텍스트 목록) != batch['label'] (개별 데이터의 라벨 id)
  batch['label_str'] = klue_tc_label.int2str(batch['label'])
  return batch

# .map(make_str_label -> 모든 개별 데이터셋에 make_str_label 적용
# batched=True -> 배치로 처리
# batch_size=1000 -> 배치 사이즈
klue_tc_train = klue_tc_train.map(make_str_label, batched=True, batch_size=1000)

klue_tc_train[0]
# {'title': '유튜브 내달 2일까지 크리에이터 지원 공간 운영', 'label': 3, 'label_str': '생활문화'}
```

# **Trainer를 사용한 학습: (1) 준비**

```python
import torch
import numpy as np
from transformers import (
    Trainer,
    TrainingArguments,
    AutoModelForSequenceClassification,
    AutoTokenizer
)

def tokenize_function(examples):
  # tokenizer -> 사전 학습된 토크나이저 사용
  # padding -> 패딩 처리
  # truncation=True -> 문장이 너무 길면 자동으로 자름 (padding="max_length" 기준)
  return tokenizer(examples["title"], padding="max_length", truncation=True)

model_id = "klue/roberta-base"
# 분류용 모델 가져오기
# train_dataset.features['label'].names = 라벨의 이름 개수
# num_labels에 모델의 출력층 뉴런 개수(분류기) 설정 -> 모델이 5개의 카테고리 중 하나를 예측하도록 학습
model = AutoModelForSequenceClassification.from_pretrained(model_id, num_labels=len(train_dataset.features['label'].names))
tokenizer = AutoTokenizer.from_pretrained(model_id)

# 토큰화
train_dataset = train_dataset.map(tokenize_function, batched=True)
valid_dataset = valid_dataset.map(tokenize_function, batched=True)
test_dataset = test_dataset.map(tokenize_function, batched=True)
```

# **Trainer를 사용한 학습: (2) 학습 인자와 평가 함수 정의**

```python
# 모델 학습에 필요한 하이퍼파라미터 설정
training_args = TrainingArguments(
    # 학습한 모델과 로그 저장 폴더
    output_dir="./results",
    # 에포크 횟수
    num_train_epochs=1,
    # 학습 시 GPU/CPU 1개당 8개 샘플(batch)을 한 번에 처리
    per_device_train_batch_size=8,
    # 평가 시 GPU/CPU 1개당 8개 샘플(batch)을 한 번에 처리
    per_device_eval_batch_size=8,
    # 매 애포크가 끝날 때마다 평가 수행
    # 평가 시 과적합이 발생하면 멈춤

    evaluation_strategy="epoch",
    # 학습률 0.00005
    learning_rate=5e-5,
    # 허깅 페이스 허브에 자동으로 업로드 X
    push_to_hub=False,

    # W&B 비활성화
    report_to="none"
)

# 모델이 평가 데이터를 예측하면 logist(예측값)과 labels(정답)을 반환
# 들어가게 될 매개변수 eval_pred -> 튜플(logits, labels)
def compute_metrics(eval_pred):
  logits, labels = eval_pred

  # np.argmax -> 배열에서 최댓값 찾기
  # axis 옵션이 없으면 전체 배열에 최대값을 찾는다
  # axis=-1은 배열의 마지막 차원을 따라 최댓값을 고른다

  # 즉 모델이 예측한 정답 라벨 인덱스 반환
  predictions = np.argmax(logits, axis=-1)
  # (predictions == labels) -> 예측값과 정답이 맞는지 비교
  # mean() -> True(1), False(0)로 변호나 후 평균을 구함 -> 정확도 계산
  return {"accuracy": (predictions == labels).mean()}
```

# **Trainer를 사용한 학습: (3) 학습 진행**

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=valid_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

trainer.train()

trainer.evaluate(test_dataset)
```

# **Trainer를 사용하지 않는 학습: (2) 학습을 위한 데이터 준비**

```python
def make_dataloader(dataset, batch_size, shuffle=True):
  # with_format("torch") -> 데이터를 PyTorch Tensor 형태로 변환
  # PyTorch 모델 학습을 위해서는 torch.Tensor 형태로 변환해야 함
  dataset = dataset.map(tokenize_function, batched=True).with_format("torch") # 데이터셋에 토큰화 수행
  # 허깅페이스의 트랜스포머 라이브러리는 라벨 컬럼이 반드시 "labels"여야 동작하는 경우가 많아 호환성 맞춤
  dataset = dataset.rename_column("label", "labels") # 컬럼 이름 변경
  # 토큰화 후 input_ids, attention_mask로 생성되므로 모델이 예상하지 않는 추가 입력값 삭제
  dataset = dataset.remove_columns(column_names=['title']) # 불필요한 컬럼 제거

  # DataLoader -> 모델 학습 시 데이터를 배치 단위로 제공하는 역할
  return DataLoader(dataset, batch_size=batch_size, shuffle=shuffle)

# 데이터로더 만들기
train_dataloader = make_dataloader(train_dataset, batch_size=8, shuffle=True)
valid_dataloader = make_dataloader(valid_dataset, batch_size=8, shuffle=False)
test_dataloader = make_dataloader(test_dataset, batch_size=8, shuffle=False)
```

# **Trainer를 사용하지 않는 학습: (3) 학습을 위한 함수 정의**

```python
def train_epoch(model, data_loader, optimizer):

  # optimizer
  # 모델의 가중치를 업데이트하는 최적화 알고리즘 (Adam, AdamW 등)

  # 모델을 학습 모드로 전환
  # 학습 모드가 되면 Dropout 같은 학습용 기능 활성화
  model.train()
  total_loss = 0

  # tqdm -> data_loader를 감싸서 진행률을 추가한 이터레이터 반환
  for batch in tqdm(data_loader):
    # 기울기 -> 손실 함수의 변화를 나타내는 값
    # 경사 하강법의 이전 배치에서 계산된 기울기를 초기화
    # PyTorch는 기본적으로 기울기를 누적하기 떄문에 초기화가 필요하다
    optimizer.zero_grad()

    # 배치 데이터의 각 항목을 뽑아 GPU로 이동
    input_ids = batch['input_ids'].to(device) # 모델에 입력할 토큰 아이디
    attention_mask = batch['attention_mask'].to(device) # 모델에 입력할 어텐션 마스크
    labels = batch['labels'].to(device) # 모델에 입력할 레이블

    # 순전파 과정
    # 순전파 -> 입력 데이터를 이용해 예측값 계산
    outputs = model(input_ids, attention_mask=attention_mask, labels=labels) # 모델 계산

    # 손실 계산
    # 정답이 1, 예측이 0.8이면 손실이 작고 정답이 1, 예측이 0.2면 손실이큼
    loss = outputs.loss # 손실

    # 역전파 과정
    # 손실을 기반으로 모델의 가중치와 기울기를 계산
    # 역전파를 수행하여 모델이 학습할 방향을 알게 됨
    # 오차를 줄이기 위해 가중치를 수정하는 과정
    loss.backward() # 역전파

    # 가중치를 업데이트한다
    optimizer.step() # 모델 업데이트

    # loss.item() -> loss는 기본적으로 torch.Tensor 형식 -> item()을 사용하여 실제 숫자로 변환
    total_loss += loss.item()

  # 평균 손실 계산
  avg_loss = total_loss / len(data_loader)

  return avg_loss
```

# **Trainer를 사용하지 않는 학습: (4) 평가를 위한 함수 정의**

```python
def evaluate(model, data_loader):
  # 모델을 평가 모드로 변경
  model.eval()
  total_loss = 0
  predictions = []
  true_labels = []

  # with -> 자동으로 리소스를 할당하고 해제 가능
  # torch.no_grad() -> 기울기 계산 해제
  with torch.no_grad():
    for batch in tqdm(data_loader):
      input_ids = batch['input_ids'].to(device)
      attention_mask = batch['attention_mask'].to(device)
      labels = batch['labels'].to(device)

      outputs = model(input_ids, attention_mask=attention_mask, labels=labels)

      # outputs.logits -> 모델이 예측한 로짓값 -> 확률값이 아니라 정규화되지 않은 값
      logits = outputs.logits
      loss = outputs.loss

      total_loss += loss.item()
      # 예측값을 실제 분류값으로 변환
      # 예측값이 가장 큰 값을 가져온다
      preds = torch.argmax(logits, dim=-1)

      # extend -> 리스트에 리스트 추가
      # cpu() -> GPU 연산 결과를 CPU로 가져온다
      # numpy() -> 넘파이 배열로 변환
      predictions.extend(preds.cpu().numpy())
      true_labels.extend(labels.cpu().numpy())


  avg_loss = total_loss / len(data_loader)

  # np.array(predictions) == np.array(true_labels) -> 정답과 예측값이 맞는지 True, False를 반환
  # 해당 값으로 평균을 구해 정확도를 구한다
  accuracy = np.mean(np.array(predictions) == np.array(true_labels))

  return avg_loss, accuracy
```

# **Trainer를 사용하지 않는 학습: (5) 학습 진행**

```python
num_epochs = 1
# 옵티마이저 -> 신경망의 가중치를 업데이트하여 학습을 진행하는 알고리즘
# 최신 모델에서는 AdamW가 많이 사용됨

# model.parameters() -> 모델의 학습 가능한 모든 파라미터(가중치와 편향 값)을 가져온다
# 옵티마이저가 이 파라미터들을 업데이트하도록 설정
optimizer = AdamW(model.parameters(), lr=5e-5)

# 학습 루프
for epoch in range(num_epochs):
    print(f"Epoch {epoch+1}/{num_epochs}")

    train_loss = train_epoch(model, train_dataloader, optimizer)
    print(f"Training loss: {train_loss}")

    valid_loss, valid_accuracy = evaluate(model, valid_dataloader)
    print(f"Validation loss: {valid_loss}")
    print(f"Validation accuracy: {valid_accuracy}")

# Testing
_, test_accuracy = evaluate(model, test_dataloader)
print(f"Test accuracy: {test_accuracy}") # 정확도 0.82
```
